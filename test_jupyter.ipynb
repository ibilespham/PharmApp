```r
library("shiny")
library("rbokeh")
library("htmlwidgets")
 
ui <- fluidPage(
  rbokehOutput("rbokeh")
)
 
server <- function(input, output, session) {
  output$rbokeh <- renderRbokeh({
    invalidateLater(1500, session)
    figure(plot_width = 400, plot_height = 800) %>% ly_hexbin(rnorm(10000), rnorm(10000))
    })
}
 
shinyApp(ui, server)
```
